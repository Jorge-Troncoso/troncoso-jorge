```python
# 2. Boxplot para visualización de valores atípicos
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title("Boxplot para Identificar Valores Atípicos")
plt.xticks(rotation=45)
plt.show()

# 3. Detección de valores atípicos con el método IQR
outliers = {}

for column in df.select_dtypes(include=[np.number]).columns:  # Solo columnas numéricas
    Q1 = df[column].quantile(0.25)  # Primer cuartil
    Q3 = df[column].quantile(0.75)  # Tercer cuartil
    IQR = Q3 - Q1  # Rango intercuartílico

    # Límites inferior y superior
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identificar valores atípicos
    outliers[column] = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]

    print(f"Columna: {column}")
    print(f" - Límite inferior: {lower_bound}")
    print(f" - Límite superior: {upper_bound}")
    print(f" - Valores atípicos detectados: {len(outliers[column])}")
    print()

# 4. Resumen de valores atípicos por columna
for col, outlier_values in outliers.items():
    if not outlier_values.empty:
        print(f"Valores atípicos en la columna {col}:\n{outlier_values}\n")
```

El dataset contiene una gran cantidad de valores atípicos, especialmente en SATO2, TEMPERATURA, HGT, y FR, por lo que se usara una estrategia de imputación

```python
for column in df.select_dtypes(include=[np.number]).columns:  # Solo columnas numéricas
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    median = df[column].median()

    if column in ['PAC_EDAD', 'EVA', 'HGT']:  # Reemplazar con límites
        df[column] = df[column].apply(
            lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x)
        )
    elif column in ['SATO2', 'TEMPERATURA', 'GLASGOW', 'FR', 'DIAS_HOSP']:  # Reemplazar con la mediana
        df[column] = df[column].apply(
            lambda x: median if x < lower_bound or x > upper_bound else x
        )
```

Se utiliza un enfoque combinado que permite tratar cada columna según su naturaleza y relevancia:

Usa límites para valores extremos claros y poco frecuentes.
Usa la mediana para valores atípicos frecuentes y clínicamente importantes, las ventajas de esto son que combina dos estrategias según la naturaleza de cada columna.logra mantener la información esencial de las columnas relevantes y prepara el dataset para algoritmos sensibles a valores extremos como k-means.

Ya imputado los valores atipicos como K-means utiliza distancias (como la distancia euclidiana), por lo que es importante que las características estén en escalas similares, ya que si no estan normalizadas, los atributos con mayores magnitudes dominarán el agrupamiento.

Se usara **MinMaxScaler** ya que es generalmente más adecuado para algoritmos como k-means esto debido a que mantiene los valores en un rango específico y permite que las distancias euclidianas (que utiliza k-means) no se vean dominadas por variables con mayor magnitud.

```python
# Se probaran valores de clusters de 2 a 10

# Limitar el uso a 2 hilos para optimizar recursos
# Esta función aparte de evaluar el numero de cluster adecuados, entrega un reporte de metricas y visualización

def evaluate_clusters_with_davies_bouldin(data, min_clusters=2, max_clusters=10, n_init=5):
    """
    Evalúa el Davies-Bouldin Index en un rango de clusters y determina el número óptimo.
    
    :param data: DataFrame o array con los datos escalados.
    :param min_clusters: Número mínimo de clusters a evaluar.
    :param max_clusters: Número máximo de clusters a evaluar.
    :param n_init: Número de inicializaciones de k-means.
    :return: Número óptimo de clusters y reporte de métricas.
    """
    # Lista para guardar los Davies-Bouldin Scores
    db_scores = []

    # Limitar el uso a 2 hilos para optimizar recursos
    import os
    os.environ["OMP_NUM_THREADS"] = "2"

    # Calcular Davies-Bouldin Index para cada número de clusters
    for k in range(min_clusters, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=n_init)
        labels = kmeans.fit_predict(data)
        score = davies_bouldin_score(data, labels)
        db_scores.append(score)

    # Graficar Davies-Bouldin Index vs Número de Clusters
    plt.figure(figsize=(10, 6))
    plt.plot(range(min_clusters, max_clusters + 1), db_scores, marker='o', linestyle='--')
    plt.title('Davies-Bouldin Index vs Número de Clusters', fontsize=14)
    plt.xlabel('Número de Clusters (k)', fontsize=12)
    plt.ylabel('Davies-Bouldin Index', fontsize=12)
    plt.grid(True)
    plt.show()

    # Determinar el número óptimo de clusters (mínimo Davies-Bouldin Index)
    optimal_k = np.argmin(db_scores) + min_clusters
    print(f"El número óptimo de clusters según Davies-Bouldin Index es: {optimal_k}")

    # Generar un reporte del cluster óptimo
    kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=n_init)
    labels = kmeans.fit_predict(data)
    centroids = kmeans.cluster_centers_

    print("\nReporte del Clustering:")
    print(f"Número óptimo de clusters: {optimal_k}")
    print(f"Davies-Bouldin Index (para k={optimal_k}): {db_scores[optimal_k - min_clusters]}")
    print(f"Inercia (distancia intra-clases): {kmeans.inertia_}")

    return optimal_k, kmeans, db_scores
```

```python
# Funcion para analizar centroides

def analyze_centroids(scaler, centroids, original_columns):
    """
    Analiza los centroides de los clusters, aplicando la transformación inversa
    al escalamiento para devolver los valores originales.

    :param scaler: Instancia del escalador usado (e.g., MinMaxScaler o StandardScaler).
    :param centroids: Coordenadas de los centroides en el espacio escalado.
    :param original_columns: Lista de nombres de las columnas originales del dataset.
    :return: DataFrame con los valores originales de los centroides.
    """
    # Transformar los centroides de vuelta al espacio original
    centroids_original = scaler.inverse_transform(centroids)
    
    # Crear un DataFrame con los centroides originales
    centroids_df = pd.DataFrame(centroids_original, columns=original_columns)
    
    # Agregar un identificador de cluster
    centroids_df.index.name = "Cluster"
    centroids_df.reset_index(inplace=True)
    
    print("Centroides en los valores originales:")
    print(centroids_df)
    return centroids_df
```

## 1.3 Clase k-means

```python
# Para esta sección como no se especifica en el enunciado que dataset usar, usare nuevamente kmeans2.csv
# La clase se llama KMeansCustom, e incluye metodos de preprocesamiento adecuados para k-means
# Tambien estan los metodos los metodos initialize centroids, assign clusters, update centroids y has converged
# Se usa el decorador @staticmethod para convertir un método de una clase en un método estático.

class KMeansCustom:
    def __init__(self, n_clusters=3, max_iter=100, tol=1e-4):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.tol = tol
        self.centroids = None
        self.labels_ = None
        self.intra_cluster_distance = None
        self.inter_cluster_distance = None

    # ----- Métodos de preprocesamiento -----
    @staticmethod
    def handle_missing_values(df, strategy="mean"):
        if strategy == "mean":
            return df.fillna(df.mean())
        elif strategy == "median":
            return df.fillna(df.median())
        elif strategy == "mode":
            return df.fillna(df.mode().iloc[0])
        else:
            raise ValueError("Estrategia no soportada. Use 'mean', 'median' o 'mode'.")

    @staticmethod
    def detect_outliers(df, method="IQR", factor=1.5):
        if method == "IQR":
            Q1 = df.quantile(0.25)
            Q3 = df.quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - factor * IQR
            upper_bound = Q3 + factor * IQR
            return df.where((df >= lower_bound) & (df <= upper_bound))
        else:
            raise ValueError("Método no soportado. Use 'IQR'.")

    @staticmethod
    def handle_outliers(df, method="IQR", factor=1.5, impute_strategy="mean"):
        """
        Detecta y maneja valores atípicos en el dataset.
        Si se detectan valores atípicos, los imputa automáticamente con la estrategia dada.
        
        :param df: DataFrame con los datos.
        :param method: Método para detectar outliers ('IQR').
        :param factor: Factor de IQR para definir los outliers (default: 1.5).
        :param impute_strategy: Estrategia para imputar outliers ('mean', 'median', 'mode').
        :return: DataFrame con valores atípicos imputados.
        """
        # Detectar outliers (marcar con NaN)
        df_with_outliers = KMeansCustom.detect_outliers(df, method, factor)
        
        # Imputar los valores atípicos marcados como NaN
        df_imputed = KMeansCustom.handle_missing_values(df_with_outliers, strategy=impute_strategy)
        
        return df_imputed

    @staticmethod
    def scale_data(df):
        scaler = MinMaxScaler()
        scaled_data = scaler.fit_transform(df.values)
        return scaled_data, scaler

    # ----- Métodos principales del algoritmo k-means -----
    def initialize_centroids(self, data):
        np.random.seed(42)
        indices = np.random.choice(data.shape[0], self.n_clusters, replace=False)
        self.centroids = data[indices]
        return self.centroids

    def assign_clusters(self, data):
        distances = np.linalg.norm(data[:, np.newaxis] - self.centroids, axis=2)
        labels = np.argmin(distances, axis=1)
        return labels

    def update_centroids(self, data, labels):
        new_centroids = np.array([data[labels == k].mean(axis=0) for k in range(self.n_clusters)])
        return new_centroids

    def has_converged(self, old_centroids, new_centroids):
        return np.all(np.linalg.norm(new_centroids - old_centroids, axis=1) < self.tol)

    def fit(self, data):
        self.centroids = self.initialize_centroids(data)
        for i in range(self.max_iter):
            labels = self.assign_clusters(data)
            new_centroids = self.update_centroids(data, labels)
            if self.has_converged(self.centroids, new_centroids):
                print(f"Convergencia alcanzada en la iteración {i+1}")
                break
            self.centroids = new_centroids

        self.labels_ = labels
        self.intra_cluster_distance = self.calculate_intra_cluster_distance(data, labels)
        self.inter_cluster_distance = self.calculate_inter_cluster_distance()

    def calculate_intra_cluster_distance(self, data, labels):
        distances = [
            np.linalg.norm(data[labels == k] - self.centroids[k], axis=1).sum()
            for k in range(self.n_clusters)
        ]
        return np.sum(distances)

    def calculate_inter_cluster_distance(self):
        distances = [
            np.linalg.norm(self.centroids[i] - self.centroids[j])
            for i in range(self.n_clusters) for j in range(i + 1, self.n_clusters)
        ]
        return np.mean(distances)

    def t(self, df, k):
        """
        Método que ajusta k-means y devuelve las etiquetas de los clusters.
        """
        df = self.handle_missing_values(df, strategy="mean")
        df = self.handle_outliers(df, method="IQR", factor=1.5, impute_strategy="median")
        data_scaled, scaler = self.scale_data(df)
        self.__init__(n_clusters=k)
        self.fit(data_scaled)
        return self.labels_

# Ejemplo de uso
# Cargar dataset kmeans2.csv
df = pd.read_csv("kmeans2.csv")

# Instanciar el modelo
kmeans = KMeansCustom()

# Ajustar el modelo y obtener etiquetas
labels = kmeans.t(df, k=5)

# Mostrar resultados
print(f"Centroides:\n{kmeans.centroids}")
print(f"Etiquetas:\n{labels}")
print(f"Distancia intra clusters: {kmeans.intra_cluster_distance}")
print(f"Distancia entre clusters: {kmeans.inter_cluster_distance}")
```

El modelo k-means fue capaz de agrupar los datos en 5 clusters con una buena compactación intra-clusters (distancia intra-clusters baja). La separación entre clusters es moderada, como lo indica la distancia promedio entre centroides.

```python
# La siguiente función permite entrenar el modelo k-means con los datos de clientes de 2024, escalar los datos, 
# y guardar tanto los centroides como el escalador para su uso futuro.

def train_and_save_model(data, n_clusters, scaler_path, centroids_path):
    """
    Entrena k-means, escala los datos, y guarda el escalador y los centroides.
    :param data: DataFrame con los datos de clientes de 2024.
    :param n_clusters: Número de clusters para k-means.
    :param scaler_path: Ruta para guardar el escalador.
    :param centroids_path: Ruta para guardar los centroides.
    """
    # Escalar los datos
    scaler = MinMaxScaler()
    data_scaled = scaler.fit_transform(data)

    # Entrenar k-means
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    kmeans.fit(data_scaled)

    # Guardar el escalador y los centroides
    joblib.dump(scaler, scaler_path)
    joblib.dump(kmeans.cluster_centers_, centroids_path)
    print("Modelo entrenado y guardado.")
```

```python
# Con los centroides y el escalador almacenados, esta función permite asignar segmentos a nuevos clientes 
# sin necesidad de volver a entrenar el modelo.

def assign_segments(new_data, scaler_path, centroids_path):
    """
    Asigna segmentos a nuevos clientes utilizando centroides preentrenados.
    :param new_data: DataFrame con los datos de nuevos clientes.
    :param scaler_path: Ruta del escalador guardado.
    :param centroids_path: Ruta de los centroides guardados.
    :return: Lista con los segmentos asignados.
    """
    # Cargar el escalador y los centroides
    scaler = joblib.load(scaler_path)
    centroids = joblib.load(centroids_path)

    # Escalar los nuevos datos con el escalador entrenado
    new_data_scaled = scaler.transform(new_data)

    # Calcular distancias y asignar segmentos
    distances = np.linalg.norm(new_data_scaled[:, np.newaxis] - centroids, axis=2)
    segments = np.argmin(distances, axis=1)
    return segments
```

**Explicación de las funciones**:

Escala los datos de 2024 con MinMaxScaler.Entrena el modelo k-means con el número de clusters especificado.
Guarda el escalador y los centroides para su uso en 2025.
assign_segments:

Carga el escalador y los centroides guardados previamente. Escala los datos de los nuevos clientes con el escalador original.
Calcula la distancia de cada nuevo cliente a los centroides y asigna el segmento más cercano.

Las ventajas de esta Implementación son la reutilización del Modelo, que Los centroides preentrenados se usan para asignar segmentos sin necesidad de reentrenamiento y consistencia en el Escalamiento ya que  se garantiza que los datos de 2025 sean tratados de la misma manera que los datos de 2024.
